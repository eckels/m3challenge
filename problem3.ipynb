```python
import numpy as np  
import matplotlib.pyplot as plt  

msl = [634012, 582839, 578815, 515830, 573855, 542873, 612716, 642163, 635925, 658317, 730994, 666580, 643792, 568737, 532843, 656873, 759596, 630562, 707999, 677492, 840236, 755817, 915596, 766391, 848073, 972122, 593255, 586235, 587283, 605977, 708932, 669078, 613386, 630165, 731218, 620782, 765423, 774862]

msl.reverse()

i = 1
year = []
while i < 39:
    year.append(i)
    i = i + 1

plt.plot(year, msl, 'cd')
plt.show()
poly = np.polyfit(year, msl, 4, rcond=None, full=False, w=None, cov=False)

print(poly[0], 'x^4')
print(poly[1], 'x^3')
print(poly[2], 'x^2')
print(poly[3], 'x')
print(poly[4])

def graph(formula, x_range):  
    x = np.array(x_range)  
    y = eval(formula)
    plt.plot(x, y)
    plt.plot(year, msl, 'o')
    plt.title("Visitor Population Trend for Padre Island National Park")
    plt.ylabel("Visitor Population (People)")
    plt.xlabel("Years (years past 1979)")
    plt.show()
    

graph('poly[0]*x*x*x*x + poly[1]*x*x*x + poly[2]*x*x + poly[3]*x + poly[4]', range(0, 38))

x = 86
neweqn = poly[0]*x*x*x*x + poly[1]*x*x*x + poly[2]*x*x + poly[3]*x + poly[4]
print(neweqn)
```
